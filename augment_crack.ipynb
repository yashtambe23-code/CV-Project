```python
import cv2
import os
import numpy as np
```

```python
input_path = r"C:\Users\yasht\Desktop\dataset\images\images\crack"
output_path = r"C:\Users\yasht\Desktop\dataset\images\images\crack_aug"

os.makedirs(output_path, exist_ok=True)
```

```python
for img_name in os.listdir(input_path):
    img_path = os.path.join(input_path, img_name)
    img = cv2.imread(img_path)

    if img is None:
        continue

    base_name = img_name.split('.')[0]

    # 1. Original
    cv2.imwrite(os.path.join(output_path, base_name + "_orig.jpg"), img)

    # 2. Flip
    flip = cv2.flip(img, 1)
    cv2.imwrite(os.path.join(output_path, base_name + "_flip.jpg"), flip)

    # 3. Rotate
    h, w = img.shape[:2]
    M = cv2.getRotationMatrix2D((w/2, h/2), 15, 1)
    rotated = cv2.warpAffine(img, M, (w, h))
    cv2.imwrite(os.path.join(output_path, base_name + "_rot.jpg"), rotated)

    # 4. Brightness
    bright = cv2.convertScaleAbs(img, alpha=1.2, beta=30)
    cv2.imwrite(os.path.join(output_path, base_name + "_bright.jpg"), bright)

    # 5. Blur
    blur = cv2.GaussianBlur(img, (5,5), 0)
    cv2.imwrite(os.path.join(output_path, base_name + "_blur.jpg"), blur)

print("✅ Augmentation Done!")
```

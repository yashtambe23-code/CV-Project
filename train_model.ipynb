```python
import cv2
import numpy as np
import os
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
Cell 2: Define Dataset Path
```

```python
dataset_path = r"C:\Users\yasht\Desktop\dataset\images\images"

classes = ["inclusion", "rolled_pit", "silk_spot", "waist folding", "crack"]
```

```python
Load & Preprocess Data (CV Part)
```

```python
data = []
labels = []

for label, folder in enumerate(classes):
    folder_path = os.path.join(dataset_path, folder)
    
    for img_name in os.listdir(folder_path):
        img_path = os.path.join(folder_path, img_name)
        
        img = cv2.imread(img_path)
        if img is None:
            continue
        
        # CV preprocessing
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        blur = cv2.GaussianBlur(gray, (5,5), 0)
        edges = cv2.Canny(blur, 50, 150)
        
        resized = cv2.resize(edges, (128,128))
        
        data.append(resized)
        labels.append(label)

data = np.array(data).reshape(-1,128,128,1) / 255.0
labels = np.array(labels)

print("✅ Data Loaded:", data.shape)
```

```python
Build CNN Model
```

```python
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128,128,1)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation='relu'),
    Dense(5, activation='softmax')   # ✅ UPDATED
])
```

```python
Compile Model
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
Train Model
```

```python
model.fit(data, labels, epochs=10, batch_size=32)
```

```python
Save Model
```

```python
model.save("defect_classifier.h5")

print("✅ Model trained and saved!")
```

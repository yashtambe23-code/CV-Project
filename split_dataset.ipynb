```python
import os, shutil, random

source = r"C:\Users\yasht\Desktop\dataset\images\images"
base = r"C:\Users\yasht\Desktop\dataset_split"

train_path = os.path.join(base, "train")
test_path = os.path.join(base, "test")

os.makedirs(train_path, exist_ok=True)
os.makedirs(test_path, exist_ok=True)

split_ratio = 0.8

for class_name in os.listdir(source):
    class_folder = os.path.join(source, class_name)

    images = os.listdir(class_folder)
    random.shuffle(images)

    split = int(len(images)*split_ratio)

    train = images[:split]
    test = images[split:]

    os.makedirs(os.path.join(train_path, class_name), exist_ok=True)
    os.makedirs(os.path.join(test_path, class_name), exist_ok=True)

    for img in train:
        shutil.copy(os.path.join(class_folder,img),
                    os.path.join(train_path,class_name,img))

    for img in test:
        shutil.copy(os.path.join(class_folder,img),
                    os.path.join(test_path,class_name,img))

print("✅ Done")
```

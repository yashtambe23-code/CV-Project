```python
import zipfile
import os
```

```python
zip_path = r"C:\Users\yasht\Desktop\archive.zip"
extract_path = r"C:\Users\yasht\Desktop\dataset"
```

```python
Extract Dataset
```

```python
# Create folder if not exists
os.makedirs(extract_path, exist_ok=True)

# Extract zip
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("✅ Dataset extracted successfully!")
```

```python
verify dataset structure
```

```python
for root, dirs, files in os.walk(extract_path):
    print(f"📁 {root}")
    for d in dirs:
        print(f"   📂 {d}")
    break
```

Check Classes Inside Images Folder

```python
dataset_path = os.path.join(extract_path, "images", "images")

classes = os.listdir(dataset_path)

print("✅ Classes found:")
for c in classes:
    print("👉", c)
```
